```python
data = sns.load_dataset('iris')
data.head()
```

```python
data1 = data.iloc[:,:-1]
data1.head()
```

```python
## Elbow method to select the K value ... 
from sklearn.cluster import KMeans
```

```python
# List to store sum of squared errors for each k
sse = []

# Iterate through different values of k
for k in range(2, 21):
    kmeans = KMeans(n_clusters=k,init='k-means++', random_state=0).fit(data1)
    sse.append(kmeans.inertia_)
```

```python
# Plot the sum of squared errors for each k
plt.figure(figsize=(10, 6))
plt.plot(range(2, 21), sse,marker='o')
plt.xlabel('Number of clusters (k)')
plt.xticks([i for i in range(2,21)])
plt.ylabel('Sum of Squared Errors')
plt.title('Elbow Method for Optimal K')
plt.grid(axis='x')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3,init='k-means++')

kmeans.fit(data1)
```

```python
from sklearn.metrics import silhouette_score,silhouette_samples
```

```python
silhouette_score(data1,kmeans.labels_)
```

```python
silhouette_coefficients =[]
for k in range(2,21):
    # Fit K-Means clustering
    agglo = KMeans(n_clusters=k, random_state=42,init='k-means++').fit(data1)

    # Calculate the silhouette score
    silhouette_avg = silhouette_score(data1, agglo.labels_)
    silhouette_coefficients.append(silhouette_avg)
```

```python
silhouette_coefficients
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(2, 21), silhouette_coefficients)
plt.xticks(range(2,21))
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of Squared Errors')
plt.title('Elbow Method for Optimal K')
plt.grid(axis='x')
plt.show()
```

```python
def divisive_clustering(X, max_clusters):
    clusters = [X]
    while len(clusters) < max_clusters:
        cluster_to_split = max(clusters, key=lambda c: c.shape[0])
        kmeans = KMeans(n_clusters=2)
        labels = kmeans.fit_predict(cluster_to_split)
        clusters.remove(cluster_to_split)
        clusters.extend([
            cluster_to_split[labels == 0],
            cluster_to_split[labels == 1]
        ])
    return clusters
```
