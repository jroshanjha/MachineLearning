```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Calculate the error rate (1 - accuracy)
error_rate = 1 - accuracy_score(y_test, y_pred) ## FP+ FN / TP+TN+FP+FN
# Accuracy = ## TP+ TN / TP+TN+FP+FN
print(f"Error rate: {error_rate:.2f}")
```

```python
# Calculate other evaluation metrics
''' 
Precision is the ratio of true positives to the sum of true positives and false positives. 
It measures how many of the positive predictions made by the model are actually correct.
Precision = True Positives / (True Positives + False Positives) '''

precision = precision_score(y_test, y_pred)

''' 
Recall is the ratio of true positives to the sum of true positives and false negatives.
It measures how many of the actual positive instances in the data were correctly identified by the model.
Recall = True Positives / (True Positives + False Negatives)
'''
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
print(precision)
print(recall)
print(f1)
```

```python
''' 
The F1 score, also known as the F1 measure or the F-measure, 
is a measure of a model's performance that combines precision and recall into a single score
'''
F1 = 2 * (precision * recall) / (precision + recall)
F1
```

```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
data = pd.read_csv('data/BostonHousing.csv') 
data
```

```python
x = data.drop('medv',axis=1)
y = data['medv']
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
X_train.corr()
```

```python
import seaborn as sns 

plt.figure(figsize=(12,10))

cor = X_train.corr()

sns.heatmap(cor,annot=True,cmap=plt.cm.CMRmap_r)

plt.show()
```

```python
# With the correlation matrix, we can see that the features with high correlation are more important than the ones with low correlation. For example, the first feature is very important and corresponds to a breast shape, the second feature is then
# it will remove the first feature that is correlated with anything other feature in the dataset.
def correlation(data,threshold):
    col_corr = set()   # Set of all names of correlated columns 
    corr_matrix = data.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
    
```

```python
corr_feature = correlation(X_train,0.7) 
corr_feature
```

```python
X_train.drop(corr_feature,axis=1)
X_test.drop(corr_feature,axis=1)
```
