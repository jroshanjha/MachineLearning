```python
# Dataset divided into input and output variable 

x_input = df.drop(columns=('Exited')) # Feateure Variable ......................
y_input = df['Exited']  # Target Variable..............
```

```python
import tensorflow
import keras
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Created of ann network
ann = Sequential()
```

```python
# ann.add(Dense(units=11,activation='sigmoid')) # input layer , kernel_regularizer=L2(l2=0.01)
# #cnn.add(BatchNormalization())
# ann.add(Dropout(0.2))
# ann.add(Dense(units=7,activation='relu')) # firt input layer
# ann.add(Dropout(0.3))
# ann.add(Dense(units=6,activation='relu')) # firt input layer
# ann.add(Dropout(0.4))
# ann.add(Dense(units=1,activation='sigmoid'))

# ann.add(Dense(3,activation='sigmoid',input_dim=10)) # 11 * 3 + 3 bias 
# ann.add(Dense(1,activation='sigmoid')) # 3 * 1 + 1 bias..

ann = Sequential([
    Dense(64, activation='relu', input_shape=(10,)),  # Change this to match your input
    Dense(32, activation='relu'), # hidden layer
    Dense(1, activation='sigmoid')  # Assuming binary classification 
])
```

```python
otp = tensorflow.keras.optimizers.Adam(learning_rate = 0.01)
ann.compile(optimizer=otp,loss='binary_crossentropy',metrics=['accuracy'])
#ann.compile(loss='binary_crossentropy', optimizer='Adam') # loss='mse'
```

```python
x_train_scaled.shape,y_train.shape
```

```python
history =ann.fit(x_train_scaled, y_train, epochs=10,validation_split=0.2) # callbacks=early_stopping , validation_split=0.33,epochs=10,batch_size = 10
```

```python
ann.layers[1].get_weights()
```

```python
result = ann.predict(x_test_scaled)
result
```

```python
y_pred = np.where(result>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred) * 100
```

```python
print(classification_report(y_test,y_pred))
```
